### PARTIE 1 – Importation des bibliothèques et chargement des données

### Import des librairies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

### chargement des donnée

```python
gate_gc = pd.read_csv('C:/Users/LENOVO/Documents/Stage_DMP/port-ai-project/data/gate_gc.csv',           sep=';',  encoding='latin1')
gate_out = pd.read_csv('C:/Users/LENOVO/Documents/Stage_DMP/port-ai-project/data/gate_out.csv',         sep='\t', encoding='latin1')

print('gate_gc :', gate_gc.shape)
print('gate_out:', gate_out.shape)
```

### Fusion des dataset 

```python
df = pd.concat([gate_gc, gate_out], ignore_index=True)
print('Dataset fusionné :', df.shape)
```

### PARTIE 2 – Vérification des dimensions et aperçu des données

```python
print('--- Dataset ---')
print(df.head())
print(df.dtypes)
print()
```

```python
df.describe()
```

### PARTIE 3 – Séparation des données (Apprentissage, Validation, Test)

```python
from sklearn.model_selection import train_test_split


# Séparation des données :
# 80% pour l'apprentissage
# 20% pour validation + test
df_train, df_temp = train_test_split(df, test_size=0.2, random_state=42)

# Séparation des 20% restants en :
# 10% validation
# 10% test
df_val, df_test = train_test_split(df_temp, test_size=0.5, random_state=42)



```

### PARTIE 4 – Sauvegarde des jeux de données

```python

# Exportation des jeux de données en fichiers CSV
# Ces fichiers seront utilisés pour la suite du projet (ML)

df_train.to_csv(r"C:/Users/LENOVO/Documents/Stage_DMP/port-ai-project/data/train.csv", index=False)
df_val.to_csv(r"C:/Users/LENOVO/Documents/Stage_DMP/port-ai-project/data/validation.csv", index=False)
df_test.to_csv(r"C:/Users/LENOVO/Documents/Stage_DMP/port-ai-project/data/test.csv", index=False)
```
